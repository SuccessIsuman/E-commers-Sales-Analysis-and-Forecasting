## Decoding and encoding the dataset to 'utf-8' encoding

```python
# changing the encoding to 'utf-8' which is a standard to avoid future errors
# Define the target encoding (e.g., 'utf-8')
target_encoding = 'utf-8'

# Function to reencode a single cell
def reencode_cell(cell):
    try:
        return cell.encode(target_encoding).decode(target_encoding)
    except:
        return cell

# Apply the reencoding function to the entire DataFrame
df = df.applymap(reencode_cell)
df
```

### Extracting month from the order data so as to know sale per months
first converting the order date to a date-time object

```python
df['Order Date'].unique()
```

```python
from datetime import datetime
# due to the difference in date formating within the order date columns, 
# A for loop and a condition will be used to attack each kind of format

count=0
for i in df['Order Date']:
    x=str(i).split('/')
    if len(x)==3:
        df.loc[count,'Order Date'] = datetime.strptime(i, "%d/%m/%Y")
    else:
        df.loc[count,'Order Date'] = datetime.strptime(i, "%d-%m-%Y")
    count+=1
   
```

```python
# Extract the month and create a new column 'Month'
# write a function that extract the month from a datetime
def Month(date):
    return date.month

# Apply the function to the order column

df['Month'] = df['Order Date'].apply(Month)
```

## Testing for Stationarity

    ADF test

       Augumented Dicky fuller test

```python
### Testing For Stationarity
from statsmodels.tsa.stattools import adfuller
```

```python
test_result=adfuller(data['Sales'])
test_result
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
    
```

```python
adfuller_test(data['Sales'])
```

## Auto Regressive Model


### Final Thoughts on Autocorrelation and Partial Autocorrelation

* Identification of an AR model is often best done with the PACF.
    * For an AR model, the theoretical PACF ‚Äúshuts off‚Äù past the order of the model.  The phrase ‚Äúshuts off‚Äù means that in theory the partial autocorrelations are equal to 0 beyond that point.  Put another way, the number of non-zero partial autocorrelations gives the order of the AR model.  By the ‚Äúorder of the model‚Äù we mean the most extreme lag of x that is used as a predictor.
    
    
* Identification of an MA model is often best done with the ACF rather than the PACF.
    * For an MA model, the theoretical PACF does not shut off, but instead tapers toward 0 in some manner.  A clearer pattern for an MA model is in the ACF.  The ACF will have non-zero autocorrelations only at lags involved in the model.
    
    ARIMA p,d,q
    
    p AR model lags
    
    d differencing
    
    q MA lags

```python
# making a copy
forcast_df=data[['Sales']].copy(deep=True)
forcast_df.head()
```

### Defferentiating the sales value to make the model stationary

```python
# first differenciation
forcast_df['diff_1']=forcast_df['Sales'].diff()
# Seasonal differenciation 
forcast_df['diff_12']=forcast_df['Sales'].diff(12)
forcast_df.head(13)
```

# Training the ARIMA model

### finding the p,d,q values 

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
# ACF: Auto correlation coeffient
# PACF: partial auto correlation coeffient
# Assuming 'data' is a DataFrame with a column 'Sales_Difference' and 'Seasonal_Difference'
fig = plt.figure(figsize=(12, 8))

# Create the first subplot for autocorrelation
ax1 = fig.add_subplot(211)
fig = plot_acf(forcast_df['diff_1'].dropna(), lags=40, ax=ax1);

# Create the second subplot for partial autocorrelation
ax2 = fig.add_subplot(212)
fig = plot_pacf(forcast_df['diff_1'].dropna(), lags=15, ax=ax2)

# Display the plots
plt.show()
```

from the plot; for non seasonal data the value of p,d,q is give as below

    p = 1 or 2
    q = 1 or 3
    d = 1

```python
# importing the model
from statsmodels.tsa.arima.model import ARIMA
```

## Spliting the dataset into training and testing dataset

```python
# Spliting the dataset into 70% training set and 30% trsting set
X_train=forcast_df[:round((len(forcast_df)*70)/100)]
X_test=forcast_df[round((len(forcast_df)*70)/100)::]
X_test.head()
```

## Traing of ARIMA Model

```python
# Fit ARIMA model
order = (1, 1, 1)  # for (p, d, q)
model = ARIMA(X_train['Sales'], order=order)
trained_model = model.fit()

# checking the summary
trained_model.summary()
```

```python
# making predictions
predictions=trained_model.predict(start=X_test.index[0],end=X_test.index[-1])
predictions
```

```python
# adding the prediction to the dataset
X_test['pred']=predictions
X_test.tail(15)
```

## plotting the actual sales value and the predicted values 

```python
# Plotting the actual sales
sns.lineplot(data=X_test, x=X_test.index, y=X_test['Sales'], label='Actual Sales')

# Plotting the predicted sales
sns.lineplot(data=X_test, x=X_test.index, y=X_test['pred'], label='Predicted Sales')

# Adding legend
plt.legend()
# Rotate x-axis tick labels to be vertical
plt.xticks(rotation='vertical')
# Display the plot
plt.show()
```

Arima model performs very poorly at the prediction of the sales data

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Calculate evaluation metrics
mae_value = mean_absolute_error(X_test['Sales'], X_test['pred'])
mse_value = mean_squared_error(X_test['Sales'], X_test['pred'])
rmse_value = np.sqrt(mse_value)

# Custom function for MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_value = mean_absolute_percentage_error(X_test['Sales'], X_test['pred'])

# Print the evaluation metrics
print("Mean Absolute Error (MAE):", mae_value)
print("Mean Squared Error (MSE):", mse_value)
print("Root Mean Squared Error (RMSE):", rmse_value)
print("Mean Absolute Percentage Error (MAPE):", mape_value)
```

the mean squared error is so large signifying that ARIMA model did not learn well on the dataset
